### Symulacja Jednowymiarowego Sygnału EEG

Ponieważ nie podano konkretnego pliku, stworzę symulowany sygnał EEG, aby zademonstrować proces.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Parametry symulacji
czas_trwania_sekundy = 5  # Czas trwania sygnału w sekundach
cze_probkowania = 250  # Częstotliwość próbkowania w Hz
cze_nosna_hz = 10  # Częstotliwość 'alpha wave' w Hz

# Generowanie osi czasu
os_czasu = np.linspace(0, czas_trwania_sekundy, int(czas_trwania_sekundy * cze_probkowania), endpoint=False)

# Generowanie sygnału sinusoidalnego (fala alpha) + szum
sygnal_eeg = np.sin(2 * np.pi * cze_nosna_hz * os_czasu) + 0.5 * np.random.randn(len(os_czasu))

print(f"Wygenerowano sygnał EEG o długości: {len(sygnal_eeg)} próbek")
print(f"Częstotliwość próbkowania: {cze_probkowania} Hz")
```

### Podgląd pierwszych próbek sygnału

```python
print("Pierwsze 10 próbek sygnału EEG:")
print(sygnal_eeg[:10])
```

### Wizualizacja sygnału EEG

```python
plt.figure(figsize=(12, 4))
plt.plot(os_czasu, sygnal_eeg, lw=1)
plt.title('Symulowany Jednowymiarowy Sygnał EEG')
plt.xlabel('Czas (s)')
plt.ylabel('Amplituda')
plt.grid(True)
plt.show()
```
