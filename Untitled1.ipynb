### Symulacja Jednowymiarowego Sygnału EEG

Ponieważ nie podano konkretnego pliku, stworzę symulowany sygnał EEG, aby zademonstrować proces.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Parametry symulacji
czas_trwania_sekundy = 5  # Czas trwania sygnału w sekundach
cze_probkowania = 250  # Częstotliwość próbkowania w Hz
cze_nosna_hz = 10  # Częstotliwość 'alpha wave' w Hz

# Generowanie osi czasu
os_czasu = np.linspace(0, czas_trwania_sekundy, int(czas_trwania_sekundy * cze_probkowania), endpoint=False)

# Generowanie sygnału sinusoidalnego (fala alpha) + szum
sygnal_eeg = np.sin(2 * np.pi * cze_nosna_hz * os_czasu) + 0.5 * np.random.randn(len(os_czasu))

print(f"Wygenerowano sygnał EEG o długości: {len(sygnal_eeg)} próbek")
print(f"Częstotliwość próbkowania: {cze_probkowania} Hz")
```

### Podgląd pierwszych próbek sygnału

```python
print("Pierwsze 10 próbek sygnału EEG:")
print(sygnal_eeg[:10])
```

### Wizualizacja sygnału EEG

```python
plt.figure(figsize=(12, 4))
plt.plot(os_czasu, sygnal_eeg, lw=1)
plt.title('Symulowany Jednowymiarowy Sygnał EEG')
plt.xlabel('Czas (s)')
plt.ylabel('Amplituda')
plt.grid(True)
plt.show()
```

### Przygotowanie danych do modelu ARIMA: Podział na zestaw treningowy i testowy

Podzielimy sygnał EEG na 80% danych treningowych i 20% danych testowych. Zestaw treningowy posłuży do nauczenia modelu ARIMA, a testowy do oceny jego przewidywań.

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

# Podział danych na zestaw treningowy (80%) i testowy (20%)
rozmiar_treningowy = int(len(sygnal_eeg) * 0.8)
dane_treningowe, dane_testowe = sygnal_eeg[0:rozmiar_treningowy], sygnal_eeg[rozmiar_treningowy:]

print(f"Długość danych treningowych: {len(dane_treningowe)}")
print(f"Długość danych testowych: {len(dane_testowe)}")
```

### Dopasowanie modelu ARIMA

Teraz dopasujemy model ARIMA do danych treningowych. Wybierzemy tutaj przykładowe parametry (p,d,q) = (5,1,0). W rzeczywistych zastosowaniach należałoby przeprowadzić analizę autokorelacji i częściowej autokorelacji (ACF/PACF) lub użyć funkcji automatycznego doboru parametrów (np. `auto_arima`).

```python
# Definicja i dopasowanie modelu ARIMA
# Przykładowe parametry (p, d, q) = (5, 1, 0)
# p: liczba opóźnień obserwacji (lagów) w modelu (część AR - Autoregressive)
# d: liczba różnicowań potrzebnych do uczynienia szeregu stacjonarnym (część I - Integrated)
# q: liczba opóźnień błędu przewidywania w modelu (część MA - Moving Average)
model = ARIMA(dane_treningowe, order=(5, 1, 0))
model_fit = model.fit()

print(model_fit.summary())
```

### Generowanie przewidywań i ocena modelu

Wygenerujemy przewidywania dla zestawu testowego i porównamy je z rzeczywistymi wartościami.

```python
# Generowanie przewidywań
prognozy = model_fit.predict(start=len(dane_treningowe), end=len(sygnal_eeg)-1)

# Ocena modelu - błąd średniokwadratowy (MSE)
mse = mean_squared_error(dane_testowe, prognozy)
print(f'Mean Squared Error (MSE): {mse:.4f}')
```

### Wizualizacja przewidywań ARIMA

Na poniższym wykresie zobaczymy, jak przewidywania modelu ARIMA porównują się z rzeczywistymi danymi testowymi.

```python
# Tworzenie osi czasu dla danych testowych
os_czasu_test = os_czasu[rozmiar_treningowy:]

plt.figure(figsize=(15, 6))
plt.plot(os_czasu[rozmiar_treningowy-100:rozmiar_treningowy], sygnal_eeg[rozmiar_treningowy-100:rozmiar_treningowy], label='Dane Treningowe (ostatnie 100 próbek)', color='blue') # Ostatnie 100 próbek treningowych
plt.plot(os_czasu_test, dane_testowe, label='Rzeczywiste Dane Testowe', color='orange')
plt.plot(os_czasu_test, prognozy, label='Prognozy ARIMA', color='green', linestyle='--')
plt.title('Przewidywania Modelu ARIMA vs Rzeczywisty Sygnał EEG (zestaw testowy)')
plt.xlabel('Czas (s)')
plt.ylabel('Amplituda')
plt.legend()
plt.grid(True)
plt.show()
```
