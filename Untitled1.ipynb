```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import pandas as pd
import mne

# Load a sample EEG dataset from MNE
# This will download the data if not already present
data_path = mne.datasets.sample.data_path(verbose=False)
raw_fname = data_path / 'MEG' / 'sample' / 'sample_audvis_raw.fif'
raw = mne.io.read_raw_fif(raw_fname, preload=True, verbose=False)

# Select an EEG channel to use as the time series signal
# For simplicity, we'll pick the first EEG channel found
# If there are no EEG channels, you might need to select another type, e.g., 'MEG'
eeg_channels = mne.pick_types(raw.info, meg=False, eeg=True, exclude='bads')
if len(eeg_channels) > 0:
    # Get data from the first EEG channel
    sygnal_eeg = raw.get_data(picks=eeg_channels[0])[0]
    ch_name = raw.ch_names[eeg_channels[0]]
    print(f"Using EEG channel: {ch_name}")
else:
    # Fallback if no EEG channels are found, e.g., use a MEG channel
    print("No EEG channels found, falling back to first MEG channel if available.")
    meg_channels = mne.pick_types(raw.info, meg=True, eeg=False, exclude='bads')
    if len(meg_channels) > 0:
        sygnal_eeg = raw.get_data(picks=meg_channels[0])[0]
        ch_name = raw.ch_names[meg_channels[0]]
        print(f"Using MEG channel: {ch_name}")
    else:
        raise ValueError("No suitable EEG or MEG channels found in the dataset.")

# Adjust parameters for the new data
cze_probkowania = raw.info['sfreq']  # Sampling frequency from MNE raw object

# Generate os_czasu based on the new signal length and sampling frequency
os_czasu = np.arange(len(sygnal_eeg)) / cze_probkowania

print(f"Loaded EEG signal. Length: {len(sygnal_eeg)} samples")
print(f"Actual sampling frequency: {cze_probkowania} Hz")
```

### Przygotowanie danych do modelu ARIMA: Podział na zestaw treningowy i testowy

Podzielimy sygnał EEG na 80% danych treningowych i 20% danych testowych. Zestaw treningowy posłuży do nauczenia modelu ARIMA, a testowy do oceny jego przewidywań.

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

# Podział danych na zestaw treningowy (80%) i testowy (20%)
rozmiar_treningowy = int(len(sygnal_eeg) * 0.8)
dane_treningowe, dane_testowe = sygnal_eeg[0:rozmiar_treningowy], sygnal_eeg[rozmiar_treningowy:]

print(f"Długość danych treningowych: {len(dane_treningowe)}")
print(f"Długość danych testowych: {len(dane_testowe)}")
```

### Dopasowanie modelu ARIMA

Teraz dopasujemy model ARIMA do danych treningowych. Wybierzemy tutaj przykładowe parametry (p,d,q) = (5,1,0). W rzeczywistych zastosowaniach należałoby przeprowadzić analizę autokorelacji i częściowej autokorelacji (ACF/PACF) lub użyć funkcji automatycznego doboru parametrów (np. `auto_arima`).

```python
# Definicja i dopasowanie modelu ARIMA
# Używamy parametrów (p, d, q) = (5, 1, 0)
model = ARIMA(dane_treningowe, order=(5, 1, 0))
model_fit = model.fit() # Removed disp=False

print(model_fit.summary())
```

### Dwustopniowy Model Predykcji: SARIMAX z Korekcją Residuów

Zgodnie z sugestią, wprowadzamy dwustopniowy model predykcji. Pierwszy model (SARIMAX) będzie trenowany na mniejszej części danych treningowych, a jego błędy (residua) na pozostałej części danych treningowych będą korygowane przez drugi model (KNN oparty na Time Delay Embedding).

```python
# Podział pierwotnych danych treningowych na zestaw treningowy dla ARIMA i walidacyjny dla reszt
rozmiar_treningowy_base = int(len(dane_treningowe) * 0.3) # 30% oryginalnych danych treningowych
dane_treningowe_base = dane_treningowe[0:rozmiar_treningowy_base]
dane_walidacyjne_base = dane_treningowe[rozmiar_treningowy_base:] # Pozostałe 70% dla walidacji i reszt

print(f"Długość danych treningowych dla ARIMA: {len(dane_treningowe_base)}")
print(f"Długość danych walidacyjnych dla reszt: {len(dane_walidacyjne_base)}")

# Ponowne dopasowanie modelu ARIMA do mniejszego zestawu treningowego
model_sarimax_base = ARIMA(dane_treningowe_base, order=(5, 1, 0))
model_sarimax_base_fit = model_sarimax_base.fit() # Removed disp=False

print("\nPodsumowanie modelu ARIMA trenowanego na 30% danych treningowych:")
print(model_sarimax_base_fit.summary())

# Generowanie predykcji ARIMA na zestawie walidacyjnym (ostatnie 70% pierwotnych danych treningowych)
prognozy_sarimax_walidacja = model_sarimax_base_fit.predict(
    start=len(dane_treningowe_base),
    end=len(dane_treningowe)-1,
    dynamic=False # Używamy rzeczywistych wartości do przewidywania, aby obliczyć residua
)

# Obliczanie residuów modelu ARIMA
residua_sarimax = dane_walidacyjne_base - prognozy_sarimax_walidacja

print(f"\nIlość obliczonych residuów ARIMA: {len(residua_sarimax)}")
print("Pierwsze 10 residuów ARIMA:")
print(residua_sarimax[:10])
```

### Time Delay Embedding dla Residuów

Aby modelować residua za pomocą algorytmu KNN, przekształcimy je w wektory stanu przy użyciu techniki Time Delay Embedding (osadzanie opóźnień czasowych). Pozwoli to KNN "zobaczyć" kontekst czasowy każdego residuum.

```python
def create_time_delay_embedding(data, delay, dimension):
    X = []
    for i in range(len(data) - (dimension - 1) * delay):
        # Tworzy wektor stanu z opóźnionymi wartościami
        state_vector = [data[i + j * delay] for j in range(dimension)]
        X.append(state_vector)
    return np.array(X)

# Parametry Time Delay Embedding dla residuów
delay = 1 # Opóźnienie czasowe
dimension = 25 # Wymiar wektora stanu (odpowiada cyklowi fali)

# Zastosowanie Time Delay Embedding do residuów, ograniczając do ostatnich max_trajectory_length
X_embedded = create_time_delay_embedding(residua_sarimax[-max_trajectory_length:], delay, dimension)

print(f"Kształt wektorów stanu po Time Delay Embedding: {X_embedded.shape}")
print("Pierwsze 3 wektory stanu:")
print(X_embedded[:3])
```

```python
from sklearn.neighbors import KNeighborsRegressor

# Przygotowanie danych dla modelu KNN
# X_knn to wektory stanu w czasie t
# y_knn to różnica wektora stanu między t+1 a t (zmiana wektora stanu)

X_knn = X_embedded[:-1] # Wszystkie wektory stanu oprócz ostatniego

# Obliczanie różnic wektorów stanu (Wektor_stanu(t+1) - Wektor_stanu(t))
y_knn = X_embedded[1:] - X_embedded[:-1]

print(f"Kształt X_knn: {X_knn.shape}")
print(f"Kształt y_knn (różnice wektorów stanu): {y_knn.shape}")

# Inicjalizacja i trenowanie modelu KNN (k=1)
knn_model = KNeighborsRegressor(n_neighbors=1)
knn_model.fit(X_knn, y_knn)

print("Model KNN dla korekcji residuów został wytrenowany.")
```

### Predykcja Połączonego Modelu ARIMA z Korekcją KNN

Teraz, gdy oba modele są wytrenowane, połączymy ich predykcje. Model ARIMA dostarczy bazowej prognozy, a model KNN, trenowany na residuach, będzie korygował te prognozy, przewidując przyszłe zachowanie błędów.

```python
import numpy as np

# Define os_czasu_test, which was missing for plotting
os_czasu_test = os_czasu[rozmiar_treningowy:]

# 1. Predykcja bazowa ARIMA dla całego zestawu testowego
# Używamy get_forecast do przewidywania wartości poza zakresem danych treningowych.
arima_forecast_test = model_sarimax_base_fit.get_forecast(steps=len(dane_testowe)).predicted_mean

# 2. Iteracyjna predykcja residuów za pomocą KNN
predicted_residuals_knn = []

# Początkowy wektor stanu dla predykcji residuów pochodzi z ostatnich obserwacji w zestawie walidacyjnym
current_residual_state_vector = X_embedded[-1].copy()

for i in range(len(dane_testowe)):
    # Przewidujemy zmianę wektora stanu za pomocą modelu KNN
    delta_vector = knn_model.predict(current_residual_state_vector.reshape(1, -1))[0]

    # Obliczamy przewidywany następny wektor stanu
    predicted_next_residual_state_vector = current_residual_state_vector + delta_vector

    # Pierwszy element przewidywanego następnego wektora stanu to przewidywane residuum dla bieżącego kroku
    predicted_residual_for_step = predicted_next_residual_state_vector[0]
    predicted_residuals_knn.append(predicted_residual_for_step)

    # Aktualizujemy bieżący wektor stanu na następny przewidywany stan dla kolejnej iteracji
    current_residual_state_vector = predicted_next_residual_state_vector

predicted_residuals_knn = np.array(predicted_residuals_knn)

# 3. Łączenie prognozy ARIMA z korekcją residuów KNN
final_predictions_combined = arima_forecast_test + predicted_residuals_knn

# 4. Ocena połączonego modelu
mse_combined = mean_squared_error(dane_testowe, final_predictions_combined)
print(f"Mean Squared Error (MSE) dla połączonego modelu ARIMA+KNN: {mse_combined:.4f}")

# 5. Wizualizacja przewidywań
plt.figure(figsize=(15, 6))
# Wykres części danych treningowych dla kontekstu
plt.plot(os_czasu[rozmiar_treningowy-100:rozmiar_treningowy], sygnal_eeg[rozmiar_treningowy-100:rozmiar_treningowy], label='Dane Treningowe (ostatnie 100 próbek)', color='blue')
plt.plot(os_czasu_test, dane_testowe, label='Rzeczywiste Dane Testowe', color='orange')
plt.plot(os_czasu_test, arima_forecast_test, label='Prognozy ARIMA (bazowe)', color='red', linestyle=':')
plt.plot(os_czasu_test, final_predictions_combined, label='Prognozy ARIMA + Korekcja KNN', color='green', linestyle='--')
plt.title('Przewidywania Połączonego Modelu ARIMA + Korekcja Residuów KNN vs Rzeczywisty Sygnał EEG')
plt.xlabel('Czas (s)')
plt.ylabel('Amplituda EEG (mV)')
plt.legend()
plt.grid(True)
plt.show()
```

### Porównanie: Model Liniowy + Korekcja KNN

Teraz zaimplementujemy model liniowy jako bazowy, aby porównać jego wydajność z modelem ARIMA, zwłaszcza w kontekście korekcji residuów za pomocą KNN. Model liniowy nie zakłada konkretnej struktury autokorelacji i jest trenowany metodą najmniejszych kwadratów na opóźnionych wartościach sygnału.

```python
from sklearn.linear_model import LinearRegression

# Liczba opóźnień dla modelu liniowego (dla spójności używamy tej samej 'dimension' co dla KNN)
n_lags = dimension # dimension = 25

def create_features_targets(data, n_lags):
    X, y = [], []
    for i in range(n_lags, len(data)):
        X.append(data[i-n_lags:i])
        y.append(data[i])
    return np.array(X), np.array(y)

# --- Przygotowanie danych dla modelu liniowego ---

# Używamy całego dostępnego zestawu treningowego (100%)
X_train_lr, y_train_lr = create_features_targets(dane_treningowe, n_lags)

# Trenowanie modelu liniowego
lr_model = LinearRegression()
lr_model_fit = lr_model.fit(X_train_lr, y_train_lr)

print(f"Model regresji liniowej został wytrenowany z {n_lags} opóźnieniami.")

# --- Obliczanie residuów modelu liniowego na danych walidacyjnych ---

# Wyznaczamy indeks odpowiadający 70% długości danych treningowych
index_70 = int(len(dane_treningowe) * 0.70)

# Część danych, na której chcemy wyliczyć residua (od 70% do 100%)
dane_walidacyjne_lr_new = dane_treningowe[index_70:]

# Do predykcji od punktu 70%, potrzebujemy n_lags próbek wstecz
full_validation_series_lr = dane_treningowe[index_70 - n_lags:]

X_val_lr_features = []
for i in range(n_lags, len(full_validation_series_lr)):
    X_val_lr_features.append(full_validation_series_lr[i-n_lags:i])
X_val_lr_features = np.array(X_val_lr_features)

# Generowanie predykcji od 70% do 100%
prognozy_lr_walidacja = lr_model_fit.predict(X_val_lr_features)

# Obliczanie residuów dla ostatnich 30% danych treningowych
residua_lr = dane_walidacyjne_lr_new - prognozy_lr_walidacja

print(f"Kształt danych treningowych LR (X_train_lr): {X_train_lr.shape}")
print(f"Kształt danych walidacyjnych LR (y_val_lr): {dane_walidacyjne_lr_new.shape}")
print(f"Ilość obliczonych residuów LR: {len(residua_lr)}")
print("Pierwsze 10 residuów LR:")
print(residua_lr[:10])
```

### Time Delay Embedding dla Residuów Modelu Liniowego

Podobnie jak dla modelu ARIMA, zastosujemy Time Delay Embedding do residuów wygenerowanych przez model liniowy.

```python
# Zastosowanie Time Delay Embedding do residuów modelu liniowego, ograniczając do ostatnich max_trajectory_length
X_embedded_lr = create_time_delay_embedding(residua_lr[-max_trajectory_length:], delay, dimension)

print(f"Kształt wektorów stanu po Time Delay Embedding (dla LR): {X_embedded_lr.shape}")
print("Pierwsze 3 wektory stanu (dla LR):")
print(X_embedded_lr[:3])
```

### Trenowanie KNN dla Korekcji Residuów Modelu Liniowego

Wytrenujemy nowy model KNN na residuach z modelu liniowego.

```python
# Przygotowanie danych dla modelu KNN (z residuów LR)
X_knn_lr = X_embedded_lr[:-1]
y_knn_lr = X_embedded_lr[1:] - X_embedded_lr[:-1]

# Inicjalizacja i trenowanie modelu KNN (k=1) dla residuów LR
knn_model_lr = KNeighborsRegressor(n_neighbors=1)
knn_model_lr.fit(X_knn_lr, y_knn_lr)

print("Model KNN dla korekcji residuów modelu liniowego został wytrenowany.")
```

### Predykcja Połączonego Modelu Liniowego z Korekcją KNN

Teraz połączymy prognozy modelu liniowego z korekcją residuów KNN i ocenimy jego wydajność.

```python
import numpy as np
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor

# Parametry
n_lags = 25
delay = 1
dimension = 25

# 1. Odtworzenie modelu liniowego i danych pomocniczych jeśli ich brakuje
if 'lr_model_fit' not in locals():
    print("lr_model_fit brakowało. Trenowanie...")
    X_train_lr_tmp, y_train_lr_tmp = [], []
    for i in range(n_lags, len(dane_treningowe)):
        X_train_lr_tmp.append(dane_treningowe[i-n_lags:i])
        y_train_lr_tmp.append(dane_treningowe[i])
    lr_model_fit = LinearRegression().fit(np.array(X_train_lr_tmp), np.array(y_train_lr_tmp))

# 2. Odtworzenie embeddingu i modelu KNN jeśli ich brakuje
if 'X_embedded_lr' not in locals() or 'knn_model_lr' not in locals():
    print("Odtwarzanie embeddingu i modelu KNN dla residuów LR...")
    # Obliczanie residuów (ostatnie 30% treningu)
    index_70 = int(len(dane_treningowe) * 0.70)
    full_validation_series_lr = dane_treningowe[index_70 - n_lags:]
    X_val_lr_features = np.array([full_validation_series_lr[i-n_lags:i] for i in range(n_lags, len(full_validation_series_lr))])
    prognozy_lr_walidacja = lr_model_fit.predict(X_val_lr_features)
    residua_lr = dane_treningowe[index_70:] - prognozy_lr_walidacja

    # Embedding
    X_embedded_lr = create_time_delay_embedding(residua_lr[-max_trajectory_length:], delay, dimension)

    # KNN training
    X_knn_lr_tmp = X_embedded_lr[:-1]
    y_knn_lr_tmp = X_embedded_lr[1:] - X_embedded_lr[:-1]
    knn_model_lr = KNeighborsRegressor(n_neighbors=1).fit(X_knn_lr_tmp, y_knn_lr_tmp)

# 3. Predykcja bazowa modelu liniowego
lr_forecast_test = []
current_window = dane_treningowe[-n_lags:].copy()
for i in range(len(dane_testowe)):
    next_pred = lr_model_fit.predict(current_window.reshape(1, -1))[0]
    lr_forecast_test.append(next_pred)
    current_window = np.append(current_window[1:], next_pred)
lr_forecast_test = np.array(lr_forecast_test)

# 4. Predykcja residuów KNN
predicted_residuals_knn_lr = []
current_residual_state_vector_lr = X_embedded_lr[-1].copy()
for i in range(len(dane_testowe)):
    delta_vector_lr = knn_model_lr.predict(current_residual_state_vector_lr.reshape(1, -1))[0]
    current_residual_state_vector_lr = current_residual_state_vector_lr + delta_vector_lr
    predicted_residuals_knn_lr.append(current_residual_state_vector_lr[0])

predicted_residuals_knn_lr = np.array(predicted_residuals_knn_lr)
final_predictions_combined_lr = lr_forecast_test + predicted_residuals_knn_lr
mse_combined_lr = mean_squared_error(dane_testowe, final_predictions_combined_lr)

print(f"MSE (Liniowy+KNN): {mse_combined_lr:.4e}")

plt.figure(figsize=(15, 6))
plt.plot(os_czasu_test, dane_testowe, label='Rzeczywiste', color='orange')
plt.plot(os_czasu_test, lr_forecast_test, label='Liniowy (bazowy)', color='red', linestyle=':')
plt.plot(os_czasu_test, final_predictions_combined_lr, label='Liniowy + KNN', color='purple', linestyle='--')
plt.title(f'Sygnał EEG: Model Liniowy + KNN (Trajektoria: {max_trajectory_length} kroków)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from IPython.display import Markdown

# Retrieve current state parameters
try:
    intercept_lr = lr_model_fit.intercept_
    coefficients_lr = lr_model_fit.coef_
    coef_str = ', '.join([f'{c:.2e}' for c in coefficients_lr[:5]]) + " ..."
    mse_val = f"{mse_combined_lr:.4e}"
    p_val = n_lags
except NameError:
    intercept_lr = 0.0
    coef_str = "Model nie został jeszcze wytrenowany"
    mse_val = "N/A"
    p_val = 25

latex_formula_str = r"$$y_{t+1} = \beta_0 + \beta_1 y_t + \beta_2 y_{t-1} + \dots + \beta_p y_{t-p+1}$$"

markdown_content = f"""
### Podsumowanie: Wytrenowany Model Liniowy + KNN

Projekt został dostosowany do rzeczywistych danych EEG oraz ograniczenia trajektorii do ostatnich {max_trajectory_length} kroków.

#### Model Liniowy (Bazowy)
Model przewiduje wartość `y` w czasie `t+1` przy użyciu `p={p_val}` opóźnień:

{latex_formula_str}

**Parametry modelu:**
- **Wyraz wolny (intercept)** $\\beta_0$: `{intercept_lr:.2e}`
- **Pierwsze 5 współczynników ($\\beta_1 \\dots \\beta_5$):** `{coef_str}`

#### Wyniki Końcowe
- **Zbiór danych:** Realne sygnały EEG (kanał {ch_name})
- **Ograniczenie trajektorii:** {max_trajectory_length} kroków
- **Mean Squared Error (MSE) Liniowy+KNN:** `{mse_val}`
"""

display(Markdown(markdown_content))
```
